# Step 1-
### Begin with defining our parameters for the file

```python
# Import necessary libraries
import requests
import time
import re
from bs4 import BeautifulSoup
```

```python
# access the webpage as Chrome (as in Lecture 5)
my_headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
```

```python
# Define the movie for which reviews will be parsed and the number of review pages to parse
movie = ('borat')
pageNum = (6)
name = ('Nick_Sansone')
```

# Step 2-
### For each review in each of the requested pages, parse the given information

```python
# Initialize the list to be empty
review_data = []

# Try to get the webpage for the given movie
for k in range(1, pageNum+1):
    webpage = ('https://www.rottentomatoes.com/m/'+movie+'/reviews?type=&sort=&page='+str(k)+'/')
    
    html = None

    # try to scrape 5 times
    for i in range(5): 
        try:
            # get url content
            response = requests.get(webpage, headers = my_headers)
            # get the html content
            html = response.content
            # if we successuflly got the file, break the loop
            break 
        # if requests.get() threw an exception, i.e., the attempt to get the response failed
        except:
            print ('failed attempt #',i)
            # wait 2 secs before trying again
            time.sleep(2)

    # if we could not get the page 
    if not html:
       # couldnt get the page, print that we could not and continue to the next attempt
       print('Could not get page: ', webpage)
    else:
       # got the page, let the user know
       print('Successfully got page: ', webpage)
        
    
    # Use BeautifulSoup's functionality for our html item
    soup = BeautifulSoup(html.decode('ascii', 'ignore'), 'lxml')
    reviews = soup.findAll('div', {'class':re.compile('review_table_row')})
    
    for review in reviews:
        
        # Initialize to inputs to 'NA'
        review_critic, review_rating, review_source, review_text, review_date = 'NA', 'NA', 'NA', 'NA', 'NA'
        
        # find first 'a' with href=/critic/..., grab text, strip() it
        critic = review.find('a', {'href':re.compile('/critic/')})
        
        # if it exists, grab the critic text
        if critic:
            review_critic = critic.text.strip()
            
        # find second 'a' in <em with class=subtle critic-publication, grab text, strip() it
        sourcetxt = review.find('em', {'class':re.compile('subtle critic-publication')})
        
        # Even if there is no source, the html index exists for this; to get around this, strip the text
        # if this is not an empty string, append the source to the list; otherwise the source 
        # does not exist and it will add 'NA'
        if sourcetxt.text.strip() != '':
            review_source = sourcetxt.text.strip()
            
        # find <div with class=review_icon icon, grab text, strip() it
        rating = review.find('div', {'class':re.compile('review_icon')})
        
        # within the class='text', take the last index to get the rating
        # same logic used here for the review_source
        if rating.attrs['class'][3] != '':
            review_rating = rating.attrs['class'][3]
            
        # find <div with class=the_review, grab text, strip() it
        txt = review.find('div', {'class':re.compile('the_review')})
        
        # same logic as above
        if txt.text.strip() != '':
            review_text = txt.text.strip()
        
        # find <div with class=review_date, grab text, strip() it
        date = review.find('div', {'class':re.compile('review-date')})
        
        # same logic as above
        if date.text.strip() != '':
            review_date = date.text.strip()
            
        # add all the info to the data link
        # if some element isn't found (based on the logic above), the 'NA' string will be added
        review_data.append([review_critic, review_rating, review_source, review_text, review_date])
        
        
# Optionally print the data to ensure it was retrieved correctly
#print(review_data)
```

# Step 3-
### After parsing the data, save them in a file formatted as given

```python
with open(name+'-'+movie+'.txt', mode='w', encoding='utf-8') as f:
    for reviews in review_data:
        f.write(reviews[0]+'\t'+reviews[1]+'\t'+reviews[2]+'\t'+reviews[3]+'\t'+reviews[4]+'\n')
        
```

```python
# Can open the file back up as a "read" file to ensure the data was written correctly
with open(name+'-'+movie+'.txt', mode='r', encoding = 'utf-8') as f:
    data = f.read()
    print(data)
```
